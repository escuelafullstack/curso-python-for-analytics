```
#Creando una función sin argumentos que realiza una tarea

def sumar():
    print(50 + 5)

sumar()
```

```
#Creando una función con argumentos que realiza una tarea

def sumar(x,y):
    print(x + y)

sumar(10,60)
```

```
#Creando una función sin argumentos que realiza una tarea

def saludar():
    print("Hola, buenos días.")

saludar()
```

```
#Creando una función con argumentos que realiza una tarea

def saludar(nombre):
    print("Hola",nombre,", buenos días.")

saludar("Juana")
```

```
#Creando una función que devuelve un resultado

def es_par(numero):
  if numero % 2 == 0:
    return True
  else:
    return False

es_par(11)
```

```
#Creando una función que devuelve un arreglo

def tabla_del(numero):
  tabla = []
  for i in range(11):
    tabla.append(numero * i)
  return tabla
  
tabla_del(5)
```

```
#Usando map para llamar a una función

def duplicar(numero):
  return numero * 2

numeros = [2, 20, 40, 130]
print("Inicialmente",numeros)

print("Luego de aplicar map",list(map(duplicar, numeros)))
```

```
#Usando lambda

duplicado = lambda x : x * 2
print(duplicado(100))
```

```
#Usando map y lambda

nums = [1,2,3,4]
potencias = list(map(lambda x: x ** 2, nums))

print(nums)
print("Las potencias al cuadrado son:",potencias)
```
